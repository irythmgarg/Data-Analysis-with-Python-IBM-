# **📌 1. Total Revenue by Country**
# **Goal: Find which countries bring in the most money.**

```python
import pandas as pd
import matplotlib.pyplot as plt

df['Revenue'] = df['Quantity'] * df['UnitPrice']
country_revenue = df.groupby('Country')['Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(14, 6))
country_revenue.plot(kind='bar', color='skyblue')
plt.title('Total Revenue by Country')
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **📆 2. Monthly Sales Trend**
# **Goal: Analyze revenue trends over months.**

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Month'] = df['InvoiceDate'].dt.to_period('M')
monthly_revenue = df.groupby('Month')['Revenue'].sum()

monthly_revenue.plot(kind='line', marker='o', figsize=(12, 5))
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

📊 Observations:
Steady Revenue in First Half:

From January to August, revenue stays relatively stable with minor fluctuations, hovering between 500K and 750K.

This suggests a consistent but modest level of sales during the first two-thirds of the year.

Significant Growth from September Onward:

There's a noticeable spike starting in September, with a rapid increase in revenue.

October and November show strong upward momentum.

November marks the peak, hitting over 1.45 million in revenue.

Sudden Drop in December (2011):

A sharp decline in December (second instance of Dec on x-axis, i.e., December 2011).

Revenue drops drastically to nearly 400K, the lowest point on the chart.

Possible reasons: incomplete data for that month, holidays affecting sales, or reporting cutoff.



# **👥 3. Top 10 Customers by Revenue**

```python
top_customers = df.groupby('CustomerID')['Revenue'].sum().nlargest(10)

top_customers.plot(kind='bar', color='coral', figsize=(10, 4))
plt.title('Top 10 Customers by Revenue')
plt.xlabel('Customer ID')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

# **⏱️ 4. Hourly Purchase Pattern**
# Goal: Understand what time of day purchases happen. **bold text**

```python
df['Hour'] = df['InvoiceDate'].dt.hour
hourly_sales = df.groupby('Hour')['Revenue'].sum()

hourly_sales.plot(kind='bar', color='purple', figsize=(10, 4))
plt.title('Revenue by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Revenue')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

📊 Observations:
Peak Hours (10 AM – 3 PM):

Revenue peaks around 12 PM (noon) with values reaching nearly 1.35 million.

High sales activity is also visible between 10 AM and 3 PM, suggesting this is the prime shopping window.

Morning Buildup (8 AM – 10 AM):

Revenue starts rising from 8 AM, gradually increasing until it peaks.

Indicates customers begin shopping early, possibly checking items before or during work hours.

Drop-off After 3 PM:

Significant decline post 3 PM.

After 5 PM, revenue becomes negligible.

Suggests customers rarely shop in the late evening or night.

Almost No Activity Before 7 AM and After 8 PM:

Virtually no revenue is generated before 7 AM or after 8 PM.

This is likely due to business hours or customer availability.

# **9.Revenue by Month & Year**
## **➤ Total monthly revenue trend across years**

```python
df['Revenue'] = df['Quantity'] * df['UnitPrice']
monthly_revenue = df.groupby(['Year', 'Month'])['Revenue'].sum().reset_index()

# Visualization
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_revenue, x='Month', y='Revenue', hue='Year', marker='o')
plt.title('Monthly Revenue Trend by Year')
plt.xticks(range(1,13))
plt.grid(True)
plt.show()
›
```

📊 Month-wise Revenue Analysis
📅 January to August (Months 1–8):
Relatively Stable & Low Revenue:

Revenue ranges from ~0.48M to ~0.72M.

Minor ups and downs — a small peak in March (Month 3) and May (Month 5).

Lowest points are in February and April (~0.49M).

📅 September to November (Months 9–11):
Significant Revenue Growth:

September (Month 9) marks the start of a sharp rise.

October continues the growth.

November (Month 11) shows the highest revenue, peaking at ~1.46M.

This indicates a strong Q4 trend.

📅 December (Month 12):
Sharp Decline:

Drops drastically to ~0.43M, the lowest of all months.

Could indicate exhaustion of demand after November or some operational limitation.

# **10. 📅 : Revenue by Day of Week**

```python
weekday_revenue = df.groupby('Weekday')['Revenue'].sum().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
).reset_index()

# Visualization
plt.figure(figsize=(10,5))
sns.barplot(data=weekday_revenue, x='Weekday', y='Revenue', palette='viridis')
plt.title('Revenue by Day of Week')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

📊 Weekday-wise Revenue Analysis
🔼 Top Revenue Days:
Thursday – Highest revenue, over 2.1 million.

Tuesday – Close second, nearly 2.0 million.

Wednesday – Also performs well, approx 1.75 million.

These three days show strong mid-week performance, indicating this could be the peak operational window for sales or transactions.

⚖️ Moderate Revenue Days:
Monday and Friday – Around 1.55–1.6 million.

Fair performance, suggesting a smooth start and end to the workweek.

🔽 Low Revenue Days:
Sunday – The lowest, around 0.78 million.

Saturday is missing, likely had negligible or zero revenue (or excluded from dataset).

✅ Conclusions:
Peak revenue occurs mid-week, especially Thursday.

Revenue dips sharply over the weekend, with Sunday contributing the least.

This pattern might reflect:

B2B nature (less weekend activity),

Customer behavior (most active during weekdays),

Or limited weekend operations.

# 11. 📆: Top Revenue Generating **Dates**

```python
daily_revenue = df.groupby('Date')['Revenue'].sum().reset_index()

# Plot top 10 dates
top_days = daily_revenue.sort_values('Revenue', ascending=False).head(10)

plt.figure(figsize=(12,5))
sns.barplot(data=top_days, x='Date', y='Revenue', palette='coolwarm')
plt.title('Top 10 Revenue Generating Dates')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

# 12. 📆: Least Revenue Generating **Dates**

```python
daily_revenue = df.groupby('Date')['Revenue'].sum().reset_index()

# Plot top 10 dates
top_days = daily_revenue.sort_values('Revenue', ascending=False).tail(10)

plt.figure(figsize=(12,5))
sns.barplot(data=top_days, x='Date', y='Revenue', palette='coolwarm')
plt.title('Least 10 Revenue Generating Dates')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
